```python


#%% no.1: read in datasets from mat file
import scipy.io
import numpy as np
# import os
# os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
from matplotlib import pyplot as plt


mat = scipy.io.loadmat('/content/path001_574.mat')
del mat["__globals__"]
del mat["__header__"]
del mat["__version__"]


throws_length=[]
life_paths_data={}
life_lengths = []


#-------------------data cleansing-----------------------#

max_throw_length=np.zeros((574,))
life_lengths = np.zeros((574,))

for j,value in enumerate(mat.values()):
    if j <= 106:
        life_lengths[j] = value.shape[-1]
        throws_length = value[1,0:-1]
        max_throw_length[j] = max(throws_length)

    else:
        life_lengths[j] = value.shape[-1]
        throws_length = value[2,0:-1]
        max_throw_length[j] = max(throws_length)


fig, ax = plt.subplots(1,1)
ax.hist(max_throw_length,bins=30, edgecolor='black')
font={'fontname':'Times',
      'size':12}
ax.set_xlabel('Maximum throw length (index)',**font)
ax.set_ylabel('Count (index)',**font)
plt.savefig('pp4_fig2.png', dpi=300)
plt.show()


fig, ax = plt.subplots(1,1)
ax.hist(life_lengths,bins=5, edgecolor='black')
font={'fontname':'Times',
      'size':12}
ax.set_xlabel('Number of turns through whole life (index)',**font)
ax.set_ylabel('Count (index)',**font)
plt.savefig('pp4_fig3.png', dpi=300)
plt.show()


#%% no2: visualize signal degradation

max_throw_length_among_all = 320
current_waveform=[]
current_waveforms=[]
count=0

I = 20 # range of RUL: 5 4 3 2 1 0
L = max_throw_length_among_all

for j,value in enumerate(mat.values()):
    if j <= 106:
        if 20<=value.shape[-1]<=37:
            throws_length = value[1,:]
            for i in range(len(throws_length)):
                current_waveform = value[2:max_throw_length_among_all+2,i]
                current_waveforms = np.hstack((current_waveforms,current_waveform))
            life_paths_data[count] = current_waveforms
            count=count+1
            current_waveform=[]
            current_waveforms=[]

    else:
        if 20<=value.shape[-1]<=37:
            throws_length = value[2,:]
            for i in range(len(throws_length)):
                current_waveform = value[10:max_throw_length_among_all+10,i]
                current_waveforms = np.hstack((current_waveforms,current_waveform))
            life_paths_data[count] = current_waveforms
            count=count+1
            current_waveform=[]
            current_waveforms=[]


fig, ax = plt.subplots(1,1)
plt.plot(range(len(life_paths_data[5][-I*L:-1])),life_paths_data[5][-I*L:-1])
font={'fontname':'Times',
      'size':12}
ax.set_xlabel('Time index',**font)
ax.set_ylabel('Current(A)',**font)
fig.set_size_inches(11.20,1.7)
plt.savefig('pp4_fig4.png', dpi=300)
plt.show()



#%% no3: data splits into train, validate and test

la1=[]
la2=[]
data_train_list = []
data_valid_list = []
data_test_list = []


Ntest = int(count*0.25)
Nvalid = int(count*0.15)
Ntrain = int(count-Nvalid-Ntest)
train_ind_max = Ntrain
valid_ind_max = train_ind_max + Nvalid
test_ind_max = valid_ind_max + Ntest


for i,f in enumerate(life_paths_data.values()):
    if i< train_ind_max:
        b = f[-I*L:]
        b = np.double(b)
        N = len(b)
        for i in range(I):
            la2[:L] = [i]*L
            la1 = np.hstack((la1,la2))
            la1 = sorted(la1,reverse=True)
        a = la1
        la1 = []
        la2 = []
        data_temp_train = np.vstack((a,b))
        data_temp_train = data_temp_train.T
        data_train_list.append(data_temp_train)

    elif i <valid_ind_max:
        b = f[-I*L:]
        b = np.double(b)
        N = len(b)
        for i in range(I):
            la2[:L] = [i]*L
            la1 = np.hstack((la1,la2))
            la1 = sorted(la1,reverse=True)
        a = la1
        la1 = []
        la2 = []
        data_temp_valid = np.vstack((a,b))
        data_temp_valid = data_temp_valid.T
        data_valid_list.append(data_temp_valid)
    else:
        b = f[-I*L:]
        b = np.double(b)
        N = len(b)
        for i in range(I):
            la2[:L] = [i]*L
            la1 = np.hstack((la1,la2))
            la1 = sorted(la1,reverse=True)
        a = la1
        la1 = []
        la2 = []
        data_temp_test = np.vstack((a,b))
        data_temp_test = data_temp_test.T
        data_test_list.append(data_temp_test)

data_train = np.concatenate(data_train_list, axis=0) #convert the list to np arrays
data_valid = np.concatenate(data_valid_list, axis=0)
data_test = np.concatenate(data_test_list, axis=0)


#%% no4: data re-arranged as input

data_rul = {}
l={}

for i in range(I): #i=0,1,2,3,4,5 ; I=6
    data_rul[I-i-1] = data_train[(i)*L:(i+1)*L,:]

for i in range(1,Ntrain):
    for j in range(I):
        data_rul[I-j-1] = np.vstack((data_rul[I-j-1],data_train[i*I*L+j*L:i*I*L+(j+1)*L,:]))


l[0] = np.vstack((data_rul[I-1],data_rul[I-2]))
for i in range(1,I-1):  #1,2,3,4;  I=6
    l[i] = np.vstack((l[i-1],data_rul[I-2-i]))
data_train_sorted = l[I-2]

data_rul={}
l={}

for i in range(I): #i=0,1,2,3,4,5 ; I=6
    data_rul[I-i-1] = data_valid[(i)*L:(i+1)*L,:]

for i in range(1,Nvalid):
    for j in range(I):
        data_rul[I-j-1] = np.vstack((data_rul[I-j-1],data_valid[i*I*L+j*L:i*I*L+(j+1)*L,:]))


l[0] = np.vstack((data_rul[I-1],data_rul[I-2]))
for i in range(1,I-1):  #1,2,3,4;  I=6
    l[i] = np.vstack((l[i-1],data_rul[I-2-i]))
data_valid_sorted = l[I-2]

data_rul={}
l={}

for i in range(I): #i=0,1,2,3,4,5 ; I=6
    data_rul[I-i-1] = data_test[(i)*L:(i+1)*L,:]

for i in range(1,Ntest):
    for j in range(I):
        data_rul[I-j-1] = np.vstack((data_rul[I-j-1],data_test[i*I*L+j*L:i*I*L+(j+1)*L,:]))


l[0] = np.vstack((data_rul[I-1],data_rul[I-2]))
for i in range(1,I-1):  #1,2,3,4;  I=6
    l[i] = np.vstack((l[i-1],data_rul[I-2-i]))
data_test_sorted = l[I-2]


data_train = []
data_valid = []
data_test = []
data_train = data_train_sorted
data_valid = data_valid_sorted
data_test = data_test_sorted


#%% no5: Normalize using mean and std of training


dmin=data_train.min(axis=0)
dmax=data_train.max(axis=0)
max_min = dmax - dmin
data_train  = (data_train-dmin)/max_min
data_valid  = (data_valid-dmin)/max_min
data_test  = (data_test-dmin)/max_min



#%% no6: data generator

window = L
step = 4
delay = 0
batch_size = 8

def generator(data, window, delay, min_index, max_index,
              shuffle=False, batch_size=batch_size, step=step):
    if max_index is None:
        max_index = len(data) - delay - 1
    i = min_index + window

    while 1:
        if shuffle:
            sample_ind = np.random.randint(
                    min_index, max_index//window, size=batch_size)
            rows = sample_ind*window
        else:
            if i >= max_index:
                    i = min_index + window
            rows = np.arange(i, min(i + batch_size*window, max_index), window)
            i = rows[-1]+window
        samples = np.zeros((len(rows),
                            window // step,
                            (data.shape[-1]-1)))
                            # first argument is number of samples (indepedent)
                            # second argument is the number of time stamps (1440/6)
                            # third arg is e.g. 12 (size of features)
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - window, rows[j], step)
            samples[j] = data[indices,1:] #indexing reducing dimenion but slicing doesn't
            #d_min = np.amin(data[indices,1:],axis=0)
            #samples[j] = (data[indices,1:]-d_min)/(np.amax(data[indices,1:],axis=0)-d_min)
            #d_min = np.amin(data[indices,1:],axis=0)
            targets[j] = data[rows[j]-1 + delay][0] #target is the first column
        yield samples, targets



train_gen = generator(data_train,
                      window=window,
                      delay=delay,
                      min_index=0,
                      max_index=None,
                      shuffle=True,
                      step=step,
                      batch_size=batch_size) #see what None does


val_gen = generator(data_valid,
                    window=window,
                    delay=delay,
                    min_index=0,
                    max_index=None,
                    shuffle=True,
                    step=step,
                    batch_size=batch_size)


test_gen = generator(data_test,
                    window=window,
                    delay=delay,
                    min_index=0,
                    max_index=None,
                    step=step,
                    batch_size=batch_size)


val_steps = data_valid.shape[0]//(window*batch_size)
test_steps = data_test.shape[0]//(window*batch_size)


#%%  no.7(option1)-LSTM training
from keras.models import Sequential
from keras import layers
from tensorflow.keras.optimizers import RMSprop

model = Sequential()
model.add(layers.LSTM(8,
                      input_shape=(window//step,1),
                      return_sequences=True))

model.add(layers.LSTM(4,
                      return_sequences=True))

model.add(layers.LSTM(2))

model.add(layers.Dense(1))

model.summary()

model.compile(optimizer=RMSprop(learning_rate=0.0001),
              loss='mse',
              metrics=['mae','mape'])

history=model.fit(train_gen,
                  steps_per_epoch=100,
                  epochs=100,
                  validation_data=val_gen,
                  validation_steps=val_steps)


model.save('RUL20_attemp2' + '.h5')


history_dict = history.history
loss = history_dict['loss']
val_loss = history_dict['val_loss']


data_test_for_evaluate = data_valid[:,1:].reshape((len(data_valid)//window, window, 1))
targets_test = data_valid[:,:1].reshape((len(data_valid)//window, window, 1))
sample = np.zeros((1, window//step,
                   (data_valid.shape[-1]-1)))

predicted_targets = np.zeros((len(data_test_for_evaluate),))
true_targets = np.zeros((len(data_test_for_evaluate),))

for i in range(0,len(data_test_for_evaluate)):
    true_targets[i] = targets_test[i,window-1]
target_mean = true_targets.mean(axis=0)

for i in range(0,len(data_test_for_evaluate)):
    sample[0] = data_test_for_evaluate[i,::step]
    predicted_targets[i]=model.predict(sample)

MSE = sum(abs(predicted_targets-true_targets)**2)/len(true_targets)
MAE = sum(abs(predicted_targets-true_targets))/len(true_targets)

RRSE = 100 * np.sqrt(MSE * len(true_targets) / (sum(abs(true_targets-target_mean)**2)))
RAE = 100 * MAE * len(true_targets) / sum(abs(true_targets-target_mean))

print('MSE: ', MSE)
print('MAE: ', MAE)
print('RRSE: ', RRSE)
print('RAE: ', RAE)
print('target_mean: ', target_mean)
print('len(true_targets): ', len(true_targets))
print(sum(abs(true_targets-target_mean)**2))
print(sum(abs(true_targets-target_mean))/len(true_targets))



fig, ax = plt.subplots(1,1)
font={'fontname':'Times',
      'size':12}
plt.plot(range(1,len(loss)+1),loss,'b--',label='Training loss')
plt.plot(range(1,len(loss)+1),val_loss,'r-',label='Validation loss')
ax.set_xlabel('Epochs',**font)
ax.set_ylabel('MSE',**font)
plt.legend()
plt.setp(ax.legend().texts,family='Times',size=12)
y=history.history['val_loss']
ymin = min(y)
print('MSE by formula: ', MSE, ' MSE by model: ', ymin)
plt.savefig('pp4_LSTM_training.png', dpi=300)
plt.show()


def smooth_curve(points, factor=0.8):
  smoothed_points = []
  for point in points:
    if smoothed_points:
      previous = smoothed_points[-1]
      smoothed_points.append(previous * factor + point * (1 - factor))
    else:
      smoothed_points.append(point)
  return smoothed_points


fig, ax = plt.subplots(1,1)
font={'fontname':'Times',
      'size':12}
plt.plot(range(1,len(loss)+1),smooth_curve(loss),'b--',label='Training loss')
plt.plot(range(1,len(loss)+1),smooth_curve(val_loss),'r-',label='Validation loss')
ax.set_xlabel('Epochs',**font)
ax.set_ylabel('MSE',**font)
# plt.grid(visible=True)
plt.legend()
plt.setp(ax.legend().texts,family='Times',size=12)
plt.savefig('pp4_LSTM_training_smoothed.png', dpi=300)
plt.show()



#%%  no.7(option1)-LSTM testing


data_test_for_evaluate = data_test[:,1:].reshape((len(data_test)//window, window, 1))
targets_test = data_test[:,:1].reshape((len(data_test)//window, window, 1))
sample = np.zeros((1, window//step,
                   (data_test.shape[-1]-1)))

predicted_targets = np.zeros((len(data_test_for_evaluate),))
true_targets = np.zeros((len(data_test_for_evaluate),))

for i in range(0,len(data_test_for_evaluate)):
    true_targets[i] = targets_test[i,window-1]
target_mean = true_targets.mean(axis=0)

for i in range(0,len(data_test_for_evaluate)):
    sample[0] = data_test_for_evaluate[i,::step]
    predicted_targets[i]=model.predict(sample)

MSE = sum(abs(predicted_targets-true_targets)**2)/len(true_targets)
MAE = sum(abs(predicted_targets-true_targets))/len(true_targets)

RRSE = 100 * np.sqrt(MSE * len(true_targets) / (sum(abs(true_targets-target_mean)**2)))
RAE = 100 * MAE * len(true_targets) / sum(abs(true_targets-target_mean))

print('MSE: ', MSE)
print('MAE: ', MAE)
print('RRSE: ', RRSE)
print('RAE: ', RAE)
print('target_mean: ', target_mean)
print('len(true_targets): ', len(true_targets))
print(sum(abs(true_targets-target_mean)**2))
print(sum(abs(true_targets-target_mean))/len(true_targets))



fig, ax = plt.subplots(1,1)
font={'fontname':'Times',
      'size':12}
plt.plot(range(1,len(loss)+1),loss,'b--',label='Training loss')
plt.plot(range(1,len(loss)+1),val_loss,'r-',label='Validation loss')
ax.set_xlabel('Epochs',**font)
ax.set_ylabel('MSE',**font)
plt.legend()
plt.setp(ax.legend().texts,family='Times',size=12)
y=history.history['val_loss']
ymin = min(y)
print('MSE by formula: ', MSE, ' MSE by model: ', ymin)
plt.show()


def smooth_curve(points, factor=0.8):
  smoothed_points = []
  for point in points:
    if smoothed_points:
      previous = smoothed_points[-1]
      smoothed_points.append(previous * factor + point * (1 - factor))
    else:
      smoothed_points.append(point)
  return smoothed_points


fig, ax = plt.subplots(1,1)
font={'fontname':'Times',
      'size':12}
plt.plot(range(1,len(loss)+1),smooth_curve(loss),'b--',label='Training loss')
plt.plot(range(1,len(loss)+1),smooth_curve(val_loss),'r-',label='Validation loss')
ax.set_xlabel('Epochs',**font)
ax.set_ylabel('MSE',**font)
plt.legend()
plt.setp(ax.legend().texts,family='Times',size=12)
plt.show()
```
